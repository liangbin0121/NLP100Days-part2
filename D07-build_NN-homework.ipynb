```python
model = models.resnet18()
```

```python
#for name, param in "自行填入":
#    "自行填入"
for name, param in model.named_parameters():
    print(name,param.shape)
```

### 為了使 forward propagation 加速 並降低 memory 使用量，請將所有 parameters 的requires_grad 關閉，關閉之後執行  forward propagation

```python
input_ = torch.randn(1, 3, 128, 128)
```

```python
for param in model.parameters():
    #"自行填入"
    param.requires_grad=False
```

```python
output = model(input_)
print(output.shape)
```

## 第二部分：依照指令，以較簡潔的方式搭建出模型

* input_shape = torch.Size([10, 12])
* 先經過一層 nn.Linear(12, 10)
* 經過10層 nn.Linear(10, 10)
* 最後經過 nn.Linear(10, 3) 輸出
* 每一個 nn.Linear過完後要先經過 nn.BatchNorm1d 才能到下一層，輸出層不用


```python
#input_ = torch.randn(10,12)
## 示範
#Linear = nn.Linear(12,10)
#BN = nn.BatchNorm1d(10)
```

```python
class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        "自行填入"
        self.input_0 = nn.Sequential(
            nn.Linear(12,10),
            nn.BatchNorm1d(10))
        self.layer = nn.ModuleList()
        for i in range(10):
          self.layer.append(nn.Sequential(
            nn.Linear(10,10),
            nn.BatchNorm1d(10)))
        self.output_=nn.Linear(10,3)

    def forward(self, x):
        "自行填入"
        x = self.input_0(x)
        for module in self.layer:
          x = module(x)
        x = self.output_(x)  
        return x
```

```python
model = Model()
```
